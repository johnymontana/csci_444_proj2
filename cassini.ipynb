generate a 50 x 50 grid of the 'Ovals of Cassini' function:

f(x,y) = (x^2 + y^2 + a^2)^2 - 4 a^2x^2 - b^4

using a =  .49, and b = .5. Sample your grid from  x \in [-1,1], y \in [-1,1].


```python
def cassini(x, y, a=0.49, b=0.5):
    return (x**2 + y**2 + a**2)**2 - (4*a**2*x**2) - b**4
```
