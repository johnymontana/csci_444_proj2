generate a 50 x 50 grid of the 'Ovals of Cassini' function:

f(x,y) = (x^2 + y^2 + a^2)^2 - 4 a^2x^2 - b^4

using a =  .49, and b = .5. Sample your grid from  x \in [-1,1], y \in [-1,1].


```python
def cassini(x, y, a=0.49, b=0.5):
    return (x**2 + y**2 + a**2)**2 - (4*a**2*x**2) - b**4
```

```python
def get_fraction(from_value, to_value, level):
    if (to_value == from_value):
        return 0
    else:
        return ((level - from_value) / (to_value - from_value))

def marching_squares(grid, level):
    if grid.shape[0] < 2 or grid.shape[1] < 2:
        raise ValueError("Input must be at least 2x2")
        
    point_list = []
    coords = []
    coords.append(0)
    coords.append(0)
    coords[0] = 0
    coords[1] = 0
    num_steps = (grid.shape[0] - 1) * (grid.shape[1] - 1)
    
    square_case = 0
    
    for n in range(num_steps):
        
        r0, c0 = coords[0], coords[1]
        r1, c1 = r0 + 1, c0 + 1
        ul = grid[r0, c0]
        ur = grid[r0, c1]
        ll = grid[r1, c0]
        lr = grid[r1, c1]
        
        if coords[1] < grid.shape[1] - 2:
            coords[1] += 1
        else:
            coords[0] += 1
            coords[1] = 0
            
        square_case = 0
        
        if (ul > level): 
            square_case += 1
        if (ur > level):
            square_case += 2
        if (ll > level):
            square_case += 4
        if (lr > level):
            square_case += 8
            
        if (square_case != 0 and square_case !=15):
            top = [r0, c0 + get_fraction(ul, ur, level)]
            bottom = [r1, c0 + get_fraction(ll, lr, level)]
            left = [r0 + get_fraction(ul, ll, level), c0]
            right = [r0 + get_fraction(ur, lr, level), c1]
            
            if (square_case == 1):
                point_list.append(top)
                point_list.append(left)
            elif (square_case == 2):
                point_list.append(right)
                point_list.append(top)
            elif (square_case == 3):
                point_list.append(right)
                point_list.append(left)
            elif (square_case == 4):
                point_list.append(left)
                point_list.append(bottom)
            elif (square_case == 5):
                point_list.append(top)
                point_list.append(bottom)
            elif (square_case == 6):
                point_list.append(right)
                point_list.append(top)
                point_list.append(left)
                point_list.append(bottom)
            elif (square_case == 7):
                point_list.append(right)
                point_list.append(bottom)
            elif (square_case == 8):
                point_list.append(bottom)
                point_list.append(right)
            elif (square_case == 9):
                point_list.append(top)
                point_list.append(left)
                point_list.append(bottom)
                point_list.append(right)
            elif (square_case == 10):
                point_list.append(bottom)
                point_list.append(top)
            elif (square_case == 11):
                point_list.append(bottom)
                point_list.append(left)
            elif (square_case == 12):
                point_list.append(left)
                point_list.append(right)
            elif (square_case == 13):
                point_list.append(top)
                point_list.append(right)
            elif (square_case == 14):
                point_list.append(left)
                point_list.append(top)
    return point_list
            
```
